```python
## Load Data
from sklearn import datasets
iris = datasets.load_iris()
features = iris.data # iris.feature_names
target = iris.target # iris.target_names
```

### Line Chart

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.plot([0.5,2.5,4.5,6.4],[6,8,2,11], ":g")
# These plt.plot() can be added to other kinds of plots too
plt.xlabel("random numbers")
plt.title("plotting line chart")
plt.ylabel("some Y")
plt.savefig("linecharg_highres.png",dpi=400)
```

### Histogram

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import hist
%matplotlib inline
hist(features[:,0], bins=10)
hist(features[:,2],bins =[0,1,1.5,2,3,4,5,6,7], alpha = 0.5) 
# alpha is transparency
# bins can be number of bins or edges for each bin
plt.xlabel("some numerical axis")
plt.ylabel("histogram")
plt.grid(axis='y',alpha=0.5)
plt.text(3, 20, 'some text')
plt.savefig("test_histogram.png", dpi=400, transparent=True)
# plt.show() # To force display image here
```

### Scatterplot

```python
from matplotlib.pyplot import scatter
from matplotlib import pyplot as plt
import numpy as np
# To set the size/dimensions of the image in inches
plt.rcParams["figure.figsize"] = (10,3) 
scatter(features[:,0],features[:,1], s= features[:,2]**2,c=target)
# s is a vector to specify the square of size
# c is a vector to specify color
plt.savefig("scatterplot.png",dpi=300)

# To reset to default
plt.rcParams.update(plt.rcParamsDefault)
```

### Swarmplot

```python
import seaborn as sns
iris = sns.load_dataset("iris")
print(f"The loaded data is of type {type(iris)}")
sns.swarmplot(x="species",y="sepal_length", data=iris)
plt.grid(alpha=0.5)
plt.show()
```

```python
# Import necessarily libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
titanic = sns.load_dataset("titanic")

# Set up a factorplot
g = sns.factorplot("class", "survived", "sex", data=titanic, kind="bar", palette="muted", legend=False)
# The ci are the confidence intervals - can take "sd", "None". Do not have clarity on how the default CI is computed
# Remember that here aggregation takes place automatically. There are several records for a given value of 'class', 'sex', 'survived' and the value in 'survived' is averaged (mean)
                   
# Show plot
plt.show()
```

```python
# bokeh basics
from bokeh.plotting import figure
from bokeh.io import show, output_notebook

# Create a blank figure with labels
p = figure(plot_width = 600, plot_height = 600, 
           title = 'Example Glyphs',
           x_axis_label = 'X', y_axis_label = 'Y')

# Example data
squares_x = [1, 3, 4, 5]
squares_y = [8, 7, 3, 1]
circles_x = [9, 12, 4, 3, 15]
circles_y = [8, 4, 11, 6, 10]

# Add squares glyph
p.square(squares_x, squares_y, size = 12, color = 'navy', alpha = 0.6)
# Add circle glyph
p.circle(circles_x, circles_y, size = 12, color = 'red')

# Set to output the plot in the notebook
output_notebook()
# Show the plot
show(p)
```
